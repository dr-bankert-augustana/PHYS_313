<font size = 5> <b> 2.1 What is Thermodynamics? </b> </font>

<p align = 'justify'>
<a href = #Definition-Thermodynamics><b>Thermodynamics</b></a> is the study of energy, its forms, transformations, and how it influences the observable world. Despite its depth, the entire field is built on just a few foundational principles known as the "<a href = #Definition-Laws><b>Laws of Thermodynamics</b></a>"
</p>

  <br>

  * <p align = 'justify'><b>Zeroth Law of Thermodynamics:</b>
  If two <a href = #Definition-System><b>systems</b></a> are each in <a href = #Definition-Python><b>thermal equilibrium</b></a> with a third system, then they are in thermal equilibrium with each other.
  </p>

  <br>

  * <p align = 'justify'><b>First Law of Thermodynamics:</b> Any change in a system's total energy occur across its <a href = #Definition-Boundary><b>boundary</b></a> according to the principle of the <a href = #Definition-COE><b>conservation of energy</b></a>.</p>

  <br>

  * <p align = 'justify'><b>Second Law of Thermodynamics:</b> During any natural thermodynamic process, the total disorder (or <a href = #Definition-Entropy><b>entropy</b></a>) of a system and its surroundings must increase with time.

    ><b>Note:</b> This law gives us the direction of processes and is often referred to as "time's arrow."

  <br>

  * <p align = 'justify'><b>Third Law of Thermodynamics:</b> The entropy of a system approaches a time-constant value as the system's <a href = #Definition-Temperature><b>temperature</b></a> approaches <a href = #Definition-Zero><b>absolute zero</b></a>. For a pure crystalline substance, this time-constant value is vanishingly small.</p>

<font size = 5> <b> 3.1 What are Systems? </b> </font>

<p align = 'justify'>
A system is a defined quantity of matter, a group of objects, or a region in space selected for study.
</p>

<br>

><b>Note:</b><p align = 'justify'>
>The choice of system is arbitrary, but a smart choice of system can make solving a particular problem much easier.
</p>

<br>

<p align = 'justify'>
Everything <b>outside</b> the system is called the <a href = #Definition-Surroundings><b>surroundings</b></a>.
</p>

<br>

><b>Note:</b><p align = 'justify'>
>Changing the system choice changes the surroundings. What's part of the system in one setup might be part of the surroundings in another and vice-versa.
</p>

<br>

The interface between the system and its surroundings is the boundary.

<br>

><b>Note:</b><p align = 'justify'>
>The boundary may be real or imaginary, and it can be either <b>fixed</b> or <b>movable</b> as the system evolves.
</p>

<font size = 5> <b> 3.2 Types of Systems </b> </font>

We classify systems based on how mass and energy can move across their boundaries:

<br>

1. <b>Open Systems:</b>

    <p align = 'justify'>
    An <a href = #Definition-Open-System><b>open system</b></a> is a type of system whose boundary allows the <b>free movement of mass and energy</b>.
  
    ><b>Note:</b><p align = 'justify'>
    If the size, shape, and position of an open system are constant in time, then it is often referred to as a <b>control volume</b>.
    </p>

    <center>
    <div>
    <img src="https://raw.githubusercontent.com/dr-bankert-augustana/PHYS_313/refs/heads/main/Images/Module_1/Open_System.png" width="700"/>
    </div>
    </center>

    <br>

2. <b>Closed Systems:</b>

    <p align = 'justify'>
    A <a href = #Definition-Closed-System><b>closed system</b></a> is a type of system whose boundary allows the free movement of energy <b> but prohibits the exchange of mass</b>.
    </p>

    ><b>Note:</b><p align = 'justify'>
    Since no mass can enter or leave the system, the total system mass is a constant in time, hence they are sometimes referred to as a <b>control mass</b>.
    </p>

    <center>
    <div>
    <img src="https://raw.githubusercontent.com/dr-bankert-augustana/PHYS_313/refs/heads/main/Images/Module_1/Closed_System.png" width="350"/>
    </div>
    </center>

<br>

3. <b>Isolated Systems:</b>

    <p align = 'justify'>
    An <a href = #Definition-Isolated-System><b>isolated system</b></a> is a type of system whose boundary <b> prohibits the exchange of mass and energy</b>.
    </p>

<font size = 5> <b> 3.3 Properties of Systems </b> </font>

We call any measurable property of a system a <a href = #Definition-Property><b>property</b></a>. Properties are considered to be either <a href = #Definition-Intensive><b>intensive</b></a> or <a href = #Definition-Extensive><b>extensive</b></a>

<br>

  * <b>Intensive Properties</b> are those measurable qualities of the system that are <b>independent</b> of the system's size. Examples include:

    * temperature
    * pressure
    * density
    * conductivity

  <br>

  <center>
  <div>
  <img src="https://raw.githubusercontent.com/dr-bankert-augustana/PHYS_313/refs/heads/main/Images/Module_1/intensive.png" width="800"/>
  </div>
  </center>

  <br>

  * <b>Extensive Properties</b> are those measurable qualities of the system that are <b>dependent</b> on the system's size. Examples include:

    * mass
    * volume
    * energy
    * momentum

  <br>

  <center>
  <div>
  <img src="https://raw.githubusercontent.com/dr-bankert-augustana/PHYS_313/refs/heads/main/Images/Module_1/extensive.png" width="800"/>
  </div>
  </center>

  <br>

  * <b>Specific Properties</b> are measures of extensive properties <b>per unit mass</b> (e.g. specific volume or specific gravity).

<p align = 'justify'>
The <a href = #Definition-State><b>state</b></a> of a system is the set of properties which fully describe the current condition of the system.

<font size = 5> <b> 3.5 System Equilibrium </b> </font>

<p align = 'justify'>
 When the state of a system is unchanging over a reasonable amount of time, it is said to be in <a href = #Definition-Equilibrium><b>equilibrium</b></a>. There are many different kinds of equilibrium and a system is not said to be in <a href = #Definition-Thermo-Equilibrium><b>thermodynamic equilibrium</b></a> unless all the conditions of the relevent types of equilibrium are met.
</p>

  * <p align = 'justify'><b>Thermal Equilibrium:</b> A system is said to be in thermal equilibrium if the <a href = #Definition-Temperature><b>temperature</b></a> is the same throughout the entire system. Since there is no temperature differential, there is no mechanism driving any flow of heat energy.</p>

  <br>

  <center>
  <div>
  <img src="https://raw.githubusercontent.com/dr-bankert-augustana/PHYS_313/refs/heads/main/Images/Module_1/Thermal-Equilibrium.png" width="600"/>
  </div>
  </center>

  <br>

  * <p align = 'justify'><b>Mechanical Equilibrium:</b> A system is said to be in mechanical equilibrium if there is no time-dependent changes in <a href = #Definition-Pressure><b>pressure</b></a> at any point of the system. The pressure can vary in the system, and still be considered in mechanical equilibrium, <b>as long as the pressure doesn't change with time</b>.</p>

  <br>

  <center>
  <div>

  </center>

  <br>

  * <p align = 'justify'><b>Phase Equilibrium:</b> If a system involves two <a href = #Definition-Phase><b>phases</b></a>, it is in phase equilibrium when the mass of each phase reaches an equilibrium level and stays there.

  <br>

  * <p align = 'justify'><b>Chemical Equilibrium:</b> If a system's chemical composition does not change with time, it is condidered to be in chemical equilibrium (<b>no chemical reactions are occuring</b>).
